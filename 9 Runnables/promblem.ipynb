```python
import random


class Naklillm:
    # constructor
    def __init__(self):
        print("LLM CREATED ")

    # predict

    def predict(self, prompt):
        response_list = [
            "computer is machine.",
            "computer is used for programming.",
            "computer is used for gaming.",
        ]

        return {"response": random.choice(response_list)}
```

```python
class Nakliprompttemplate:
    # constructor
    def __init__(self, template, input_variables):
        self.template = template
        self.input_variables = input_variables

    # format
    def format(self, input_dict):
        return self.template.format(**input_dict)
```

```python
template = Nakliprompttemplate(template="what is {topic}?", input_variables=["topic"])
```

```python
prompt = template.format({"topic": "computer"})
```

```python
llm = Naklillm()
```

```python
llm.predict(prompt)
```
